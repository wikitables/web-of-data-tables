```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("cluster7.csv", sep="\t", names=["cluster", "table", "headers", "c1", "c2","relation", "ratio"], index_col=False, dtype={'table': np.str})
```

```python
df.head()
```

```python
df["pair_rel"]=df["c1"]+"#"+df["c2"]+"#"+df["relation"]
```

```python
df1=pd.DataFrame()
df1["table"]=df["table"]
df1["ratio"]=df["ratio"]
df1["pair_rel"]=df["pair_rel"]
df1.head()
```

```python
res = df1.pivot_table(index=['table'], columns=['pair_rel'],
                     values='ratio', aggfunc='first', fill_value=0)
res.head()
```

```python
res_count=pd.DataFrame()
for column in res:    
    res_count[column]=res[column].apply(lambda x: 1.0 if x>0 else 0)
res_count.head()
```

```python
dfsum=pd.DataFrame(res_count.sum(), columns=["value"])
dfsum=dfsum.sort_values("value", ascending=False)
```

```python
ax=dfsum[dfsum["value"]>100].plot.bar(rot=80)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(res)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc1', 'pc2'])
```

```python
principalDf.index =res.index
```

```python
principalDf.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3).fit(principalDf)
centroids = kmeans.cluster_centers_

%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D

labels = kmeans.predict(principalDf)
centers = kmeans.cluster_centers_
```
